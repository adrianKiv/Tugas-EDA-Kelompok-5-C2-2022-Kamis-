```python
import pandas as pd
import re

# Baca data dari 5 file Excel
file_paths = ['data banjir.xlsx', 'data kebakaran.xlsx', 'data kekeringan.xlsx', 'data tanah longsor.xlsx', 'data puting beliung.xlsx']
dfs = [pd.read_excel(file) for file in file_paths]

# Tambahkan kolom 'Bencana' ke masing-masing DataFrame untuk mengidentifikasi jenis bencana
for i, df in enumerate(dfs):
    df['Bencana'] = file_paths[i].replace('data ', '').replace('.xlsx', '')


# Gabungkan data menjadi satu DataFrame
combined_data = pd.concat(dfs)

def replace_comma_with_dot(value):
    if re.match(r'^\d{1,3}(,\d{3})*(\.\d+)?$', value):
        # Format benar, ganti koma dengan titik
        return value.replace(',', '.')
    else:
        # Format salah, biarkan nilai asli
        return value

combined_data['Meninggal'] = combined_data['Meninggal'].apply(replace_comma_with_dot)

# combined_data['Meninggal'] = combined_data['Meninggal'].str.replace(',', '')
# combined_data["Meninggal"] = pd.to_numeric(combined_data["Meninggal"], errors='ignore')
combined_data
# combined_data[combined_data.Bencana == 'banjir']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# drop
# Membuat plot
plt.figure(figsize=(12, 6))
sns.barplot(data=combined_data, x='Propinsi', y='Meninggal', hue='Bencana', palette='viridis')
plt.xlabel('Provinsi')
plt.ylabel('Jumlah Korban Meninggal')
plt.title('Jumlah Korban per Provinsi dari Berbagai Bencana')
plt.xticks(rotation=90)
plt.legend(title='Bencana', loc='upper right')
plt.tight_layout()

# Menampilkan plot
plt.show()
```
